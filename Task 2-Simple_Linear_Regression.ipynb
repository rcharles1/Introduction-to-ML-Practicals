```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df = pd.read_csv('dataset/04-advertising_dataset.csv')
```

```python
df.shape
```

```python
df = df.drop('Radio', axis=1)
df = df.drop('Newspaper', axis=1)
df.head()
```

```python
df.describe()
```

```python
df.plot(x='TV', y='Sales', style='o')
plt.title('TV VS SALES')
plt.xticks(np.arange(min(df['TV']), max(df['Sales'])+1, 1.0))
plt.xlabel('TV')
plt.ylabel('Sales')
plt.show()
```

```python
x = df.iloc[:, :-1].values
y = df.iloc[:,1].values
```

```python
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size=0.50,random_state=100)
```

```python
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
print(f'b = {model.intercept_}')
print(f'w = {model.coef_}')
```

```python
all_y_pred = model.predict(x)
plt.figure(figsize=(12,6))
plt.plot(x, all_y_pred)
plt.plot(x, y,'ro')
plt.title('Actual vs Predicted')
plt.xticks(np.arange(min(x), max(x)+1, 1.0))
plt.xlabel('TV')
plt.ylabel('Sales')
```

```python
print("Mean squared error: %.2f" % mean_squared_error(y_test, y_pred))

print("Coefficient of determination: %.2f" %r2_score(y_test, y_pred))
```
